```python
salary_cleaned = salary_cleaned.drop_duplicates()
salary_cleaned.info()
```

3. hapus data yg tidak digunakan atau redundant. dalam dataframe ini kolom salary, salary currency, salary in usd merupakan sama2 gaji, sehingga redundant. untuk itu perlu dihapus

```python
salary_cleaned = salary_cleaned.drop('salary', axis = 1)
salary_cleaned = salary_cleaned.drop('salary_currency', axis=1)
```

```python
salary_cleaned
```

```python
salary_cleaned['experience_level'] = salary_cleaned['experience_level'].replace ('EN','Junior')
salary_cleaned['experience_level'] = salary_cleaned['experience_level'].replace ('MI','Intermediete')
salary_cleaned['experience_level'] = salary_cleaned['experience_level'].replace ('SE','Expert')
salary_cleaned['experience_level'] = salary_cleaned['experience_level'].replace ('EX','Director')

salary_cleaned['employment_type'] = salary_cleaned['employment_type'].replace ('PT','Part-Time')
salary_cleaned['employment_type'] = salary_cleaned['employment_type'].replace ('FT','Full-Time')
salary_cleaned['employment_type'] = salary_cleaned['employment_type'].replace ('CT','Contract')
salary_cleaned['employment_type'] = salary_cleaned['employment_type'].replace ('FL','Freelance')

salary_cleaned['company_size'] = salary_cleaned['company_size'].replace ('S','Small')
salary_cleaned['company_size'] = salary_cleaned['company_size'].replace ('M','Medium')
salary_cleaned['company_size'] = salary_cleaned['company_size'].replace ('L','Large')

salary_cleaned['remote_ratio'] = salary_cleaned['remote_ratio'].replace (0,'WFO')
salary_cleaned['remote_ratio'] = salary_cleaned['remote_ratio'].replace (50,'hybrid')
salary_cleaned['remote_ratio'] = salary_cleaned['remote_ratio'].replace (100,'WFH')
```

```python
salary_cleaned
```

4. cek outlier

```python
salary_cleaned.boxplot(column = 'salary_in_usd')
```

```python
salary_cleaned['salary_in_usd'].describe()
```

ternyata dr hasil diatas, diketahui std dan jarak antara q3 ke nilai max besar, jauh, shg ada outliers

```python
q1 = salary_cleaned['salary_in_usd'].quantile(0.25)
q3 = salary_cleaned['salary_in_usd'].quantile(0.75)

upper_limit = q3 + 1.5 * (q3 - q1)
lower_limit = q1 - 1.5 * (q3 - q1)
```

```python
salary_cleaned = salary_cleaned[salary_cleaned['salary_in_usd'] > lower_limit]
salary_cleaned = salary_cleaned[salary_cleaned['salary_in_usd'] < upper_limit]
salary_cleaned['salary_in_usd'].describe()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.bar(['job_title']['salary_in_usd'].average().sort_values(ascending=False), color = 'green')
plt.xlabel('job title', fontsize=15)
plt.ylabel('salary', fontsize=15)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/10000000).astype(int))
plt.show()
```

```python
salary_cleaned.groupby('job_title')['salary_in_usd'].average().sort_values(ascending=False).plot(kind='bar', color='green')
plt.title('distribution salary on tech job', loc='center', pad=30, fontsize=15, color='blue')
plt.xlabel('job title', fontsize=15)
plt.ylabel('salary', fontsize=15)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/10000000).astype(int))
plt.xticks(rotation=0)
plt.show()
```

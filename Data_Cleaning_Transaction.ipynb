setelah dicek, maka diketahui ada:

1. cek data null

```python
Transaction.isna()
```

klu false brarti g ada null. Untuk melihat scr keseluruhan maka

```python
Transaction.isna().sum()
```

filter data kolom yg null

```python
Transaction[Transaction['Individual_Price_US$'].isna()]
```

```python
Transaction[Transaction['Amount US$'].isna()]
```

setelah dicek, ternyata cuma 12 data yg null dari 2000 data, shg tdk smp 1%. maka data yg null bisa dihapus. Untuk menghapus menggunakan tanda ~

```python
Transaction_without_null = Transaction[~Transaction['Individual_Price_US$'].isna()]
Transaction_without_null = Transaction_without_null[~Transaction_without_null['Amount US$'].isna()]
Transaction_without_null
```

2. cek tipe data

```python
Transaction_without_null.info()
```

Individual price dan amount harusnya tipe data angka atau int, setelah dilihat data kolom tsb, terlihat ada koma, koma ini merupakan karakter shg terbaca menjadi objek. untuk itu harus dihilangkn dulu komanya

```python
Transaction_without_null['Amount US$'] = Transaction_without_null['Amount US$'].str.replace(',' , '')
Transaction_without_null['Amount US$']
```

```python
Transaction_without_null['Amount US$'] = Transaction_without_null['Amount US$'].astype('int')
```

```python
Transaction_without_null['Individual_Price_US$'] = Transaction_without_null['Individual_Price_US$'].str.replace(',' , '')
Transaction_without_null['Individual_Price_US$']
```

```python
Transaction_without_null['Individual_Price_US$'] = Transaction_without_null['Individual_Price_US$'].astype('int')
```

```python
Transaction_without_null.info()
```

```python
Transaction_without_null['Date'] = pd.to_datetime(Transaction['Date'], format='%m/%d/%Y')
```

```python
Transaction_without_null['Date']
```

```python
Transaction_without_null.info()
```

```python
Transaction_without_null['Time_Transaction'] = pd.to_datetime(Transaction_without_null['Time_Transaction'], format='%H:%M:%S')
```

3. Cek Duplikat

```python
Transaction_without_null = Transaction_without_null.drop_duplicates()
```

setelah di cek, tidak ditemukan duplikasi

4. Cek outliers

```python
Transaction_without_null.boxplot(column='Individual_Price_US$')
```

```python
Transaction_without_null['Individual_Price_US$'].describe()
```

jarak antara Q3 ke maksimal terlalu jauh, shg menunjukkan adanya outliers

batas atas = Q3 + 1,5 (Q3 - Q1)
batas bawah = Q1 - 1,5 (Q3 - Q1)

```python
q1 = Transaction_without_null['Individual_Price_US$'].quantile (0.25)
q3 = Transaction_without_null['Individual_Price_US$'].quantile (0.75)
upper_limit = q3 + 1.5 * (q3 - q1)
lower_limit = q1 - 1.5 * (q3 - q1)
```

untuk memfilter, maka digunakan sintaks di bawah shg diperoleh data yg dalam rentang antara min dan max.

```python
Transaction_without_null = Transaction_without_null[Transaction_without_null['Individual_Price_US$'] > lower_limit]
Transaction_without_null = Transaction_without_null[Transaction_without_null['Individual_Price_US$'] < upper_limit]
```

```python
Transaction_without_null['Individual_Price_US$'].describe()
```

```python
Transaction_without_null.boxplot(column='Individual_Price_US$')
```
